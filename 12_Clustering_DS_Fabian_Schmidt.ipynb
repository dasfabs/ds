```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/dasfabs/ds/main/edlich-kmeans-A0.csv'
df = pd.read_csv(url, error_bad_lines=False)
df.head()

distortions = []
K = range(1,15)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df)
    distortions.append(kmeanModel.inertia_)
    
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Verzerrung')
plt.title('Elbow Method - best k')
plt.show()   
```

```python
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from numpy import random

plt.rcParams['figure.figsize'] = (16, 8)
plt.style.use('ggplot')

url = 'https://raw.githubusercontent.com/dasfabs/ds/main/edlich-kmeans-A0.csv'
data = pd.read_csv(url, error_bad_lines=False)

f1 = data['V1'].values
f2 = data['V2'].values
f3 = data['V3'].values

X = np.array(list(zip(f1, f2, f3)))
y = random.randint(3, size=(len(X)))

kmeans = KMeans(n_clusters=5)
kmeans = kmeans.fit(X)
labels = kmeans.predict(X)
C = kmeans.cluster_centers_

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c='#ff0000', s=500)
```
