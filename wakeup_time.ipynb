```python
import numpy
import matplotlib.pyplot as plot
from scipy import stats
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/dasfabs/ds/main/wecker.csv')
#df.head
x = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
print(x)
print(y)
stats.describe(y)
plot.plot(y)
plot.show()
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 2/7, random_state = 0)
print(xTrain)
print(yTrain)
```

```python
linearRegressor = LinearRegression()
linearRegressor.fit(xTrain, yTrain)
print(linearRegressor)
yPrediction = linearRegressor.predict(xTest)
print(yPrediction)
```

```python
plot.scatter(xTrain, yTrain, color = 'red')
plot.plot(xTrain,linearRegressor.predict(xTrain), color = 'blue')
plot.title('Wake up time (Training set)')
plot.xlabel('Days')
plot.ylabel('Time')
plot.show()
plot.scatter(xTest, yTest, color = 'red')
plot.plot(xTrain,linearRegressor.predict(xTrain), color = 'blue')
plot.title('Wake up time (Test set)')
plot.xlabel('Days')
plot.ylabel('Time')
plot.show()
```
